```python
import time
import matplotlib.pyplot as plt
import numpy as np
from sklearn.neighbors import KDTree, BallTree
```

Массивы для хранения времени

```python
kd_times = []
ball_times = []
kd_query_times = []
ball_query_times = []
```

```python
for i in range(1,70):
    data = np.random.rand(10000, i)
    query = np.random.rand(1, i)
    avg_timeKd=0
    avg_timeKdQuery=0
    avg_timeBall=0
    avg_timeBallQuery=0
    for j in range(100):

        start = time.time()
        kd_tree = KDTree(data, leaf_size=30)
        avg_timeKd=avg_timeKd+(time.time() - start)
        start = time.time()
        kd_tree.query(query, k=10, return_distance=False)
        avg_timeKdQuery=avg_timeKdQuery+(time.time() - start)

        start = time.time()
        ball_tree = BallTree(data, leaf_size=30)
        avg_timeBall=avg_timeBall+(time.time() - start)
        start = time.time()
        ball_tree.query(query, k=10, return_distance=False)
        avg_timeBallQuery=avg_timeBallQuery+(time.time() - start)
    kd_times.append(avg_timeKd/100)
    kd_query_times.append(avg_timeKdQuery/100)

    ball_times.append(avg_timeBall/100)
    ball_query_times.append(avg_timeBallQuery/100)
```

График затраченного времени на создание дерева KD Tree и Ball Tree

```python
plt.plot(kd_times, label="KD Tree")
plt.plot(ball_times, label="Ball Tree")
plt.xlabel("Размерность")
plt.ylabel("Время")
plt.legend()
plt.show()
```

График затраченного времени на создание запроса KD Tree и Ball Tree

```python
plt.plot(kd_query_times, label="KD Tree query")
plt.plot(ball_query_times, label="Ball Tree query")
plt.xlabel("Размерность")
plt.ylabel("Время")
plt.legend()
plt.show()
```
