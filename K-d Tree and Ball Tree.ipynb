```python
import time
import matplotlib.pyplot as plt
import numpy as np
from sklearn.neighbors import KDTree, BallTree
```

Массивы для хранения времени

```python
kd_times = []
ball_times = []
kd_query_times = []
ball_query_times = []
```

```python
for i in range(1,100):
    data = np.random.rand(10000, i)
    query = np.random.rand(1, i)

    start = time.time()
    kd_tree = KDTree(data, leaf_size=30)
    kd_times.append(time.time() - start)
    start = time.time()
    kd_tree.query(query, k=10, return_distance=False)
    kd_query_times.append(time.time() - start)

    start = time.time()
    ball_tree = BallTree(data, leaf_size=30)
    ball_times.append(time.time() - start)
    start = time.time()
    ball_tree.query(query, k=10, return_distance=False)
    ball_query_times.append(time.time() - start)
```

График затраченного времени на создание дерева KD Tree и Ball Tree

```python
plt.plot(kd_times, label="KD Tree")
plt.plot(ball_times, label="Ball Tree")
plt.legend()
plt.show()
```

График затраченного времени на создание запроса KD Tree и Ball Tree

```python
plt.plot(kd_query_times, label="KD Tree query")
plt.plot(ball_query_times, label="Ball Tree query")
plt.legend()
plt.show()
```
